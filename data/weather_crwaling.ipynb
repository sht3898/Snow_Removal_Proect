# pandas 형태로 날씨 2017~2019 날씨 정보 저장

```python
weather_columns = ('year', 'month', 'avg_temp', 'high_temp', 'low_temp', 'avg_cloud', 'daily_rain')
date = []
info = []

for year in (2017, 2018, 2019):
    for month in (1, 2, 3, 11, 12):
        
        driver.get(f'https://www.weather.go.kr/weather/climate/past_cal.jsp?stn=133&x=23&y=12&yy={year}&mm={month}&obs=1')
        days = driver.find_elements_by_class_name('align_left')
        for d in days:
            if 0 < len(d.text) <= 3 and d.text != ' ':
                day = d.text
                date.append(day[:-1])
            else:
                if d.text != ' ':
                    tmp = d.text.split('\n')
                    avg_temp = tmp[0][5:-1]
                    high_temp = tmp[1][5:-1]
                    low_temp = tmp[2][5:-1]
                    avg_cloud = tmp[3][5:]
                    daily_rain = tmp[4][5:-2]
                    info.append([year, month, avg_temp, high_temp, low_temp, avg_cloud, daily_rain])
```

```python
date_df = pd.DataFrame(date)
info_df = pd.DataFrame(info, columns=weather_columns)
# 데이터 추가
info_df['day'] = date_df
# 컬럼 순서 변경
weather_df = info_df[['year', 'month', 'day', 'avg_temp', 'high_temp', 'low_temp', 'avg_cloud', 'daily_rain']]
```

```python
# csv 파일로 저장
weather_df.to_csv("weather_data.csv", mode='w')
```

# 2017~2019 기상 상태 정보 저장

```python
driver = webdriver.Chrome('./chromedriver')
```

## 한개로 테스트

```python
weather_columns = ('year', 'month', 'day', 'status')
date = []
info = []
driver.get('https://www.weather.go.kr/weather/climate/past_cal.jsp?stn=133&yy=2017&x=31&y=10&mm=1&obs=9')
days = driver.find_elements_by_class_name('align_left')
for d in days:
    if d.text.endswith('일'):
        last = d.text
        date.append(d.text[:-1])
    else:
        if last.endswith('일'):
            info.append(d.text.split('\n'))
info = info[:len(date)]
```

## 제대로 만들기

```python
weather_columns = ('year', 'month', 'status')
date = []
info = []
for year in (2017, 2018, 2019):
    for month in (1,2,3,11,12):
        temp_date = []
        temp_info = []
        driver.get(f'https://www.weather.go.kr/weather/climate/past_cal.jsp?stn=133&yy={year}&x=31&y=10&mm={month}&obs=9')
        days = driver.find_elements_by_class_name('align_left')
        for d in days:
            if d.text.endswith('일'):
                last = d.text
                temp_date.append(d.text[:-1])
            else:
                if last.endswith('일'):
                    temp_info.append([year, month, ' '.join(d.text.split('\n'))])
        temp_info = temp_info[:len(temp_date)]
        date.extend(temp_date)
        info.extend(temp_info)
```

```python
date_df = pd.DataFrame(date)
info_df = pd.DataFrame(info, columns=weather_columns)
```

```python
info_df['day'] = date_df
weather_df = info_df[['year', 'month', 'day', 'status']]
```

```python
weather_df.to_csv('weather_status_data.csv', mode='w')
```
